```python
# Drop the sign of $ and B
billionaires['NetWorth']=billionaires['NetWorth'].str.strip("$")
billionaires['NetWorth']=billionaires['NetWorth'].str.strip("B")
billionaires.head()
```

```python
# Top 10 Billbillionaires as per NetWorth
data=billionaires.sort_values(by=['NetWorth'],ascending=False).head(10)
plt.figure(figsize=(20,5))
plt.bar(data['Name'],data['NetWorth'])
plt.show()
```

## Top 5 Domain with the most Number of Billionaires

```python
a=billionaires["Source"].value_counts().head(5)
index=a.index
print(index)
sources=a.values
print(sources)
col=["red","blue","skyblue","green","tomato"]
plt.figure(figsize=(5,5))
plt.pie(sources,labels=index,colors=col)
center=plt.Circle((0,0),0.5,color='white')
fig=plt.gcf()
fig.gca().add_artist(center)
plt.rc('font',size=12)
plt.title("top 5 domains with the most number of billionaires",fontsize=30)
plt.show()
```
